## Formula para calcular area
$A= \pi * r^2$
---
## Formula para calucular el perimetro
$P= 2 * \pi * r$
---
## Datos de entrada
$r=?$

---
## Diagrama de flujo en lucidchart
![Logo de Markdown](https://raw.githubusercontent.com/JacoboValenzuela128/EPPL-rea-y-per-metro-de-un-circulo-mis-im-genes/refs/heads/main/DF%20Area%20perimetro%20circulo.png)

```python
# Calcular area y perimetro de un circulo
# Variables A, P, r
# Formulas A = pi * r^2 y P = 2 * pi * r
# Capturar el valor de r
import math
print("Calcular el area y perimetro de un circulo")
r = int(input("Ingrese el valor de r: "))
A = math.pi * (r ** 2)
P = 2 * math.pi * r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Arath Valenzuela NL:52")
```
